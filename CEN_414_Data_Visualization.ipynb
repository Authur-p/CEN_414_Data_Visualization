                    Measures of dispersion 
with aid of graphs, i would visualize the measures of dispersion e.g mean, standard deviation, etc of GDP in African countries.

```python
plt.figure(figsize=(10,10))

def interquartile_range(sample):
    q3, q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

def quartile_range(sample):
    q3, q1 = np.percentile(sample, [75 ,25])
    return (q3 - q1)/2

def _range(sample):
    return sample.max() - sample.min()

def create_plot(function, label):
    dispersion_list = []
    for year in range(2010,2020):
            dispersion_list.append(function(data[data["Year"] == year].Value))
              
    plt.plot(range(2010,2020), dispersion_list, label=label)
    plt.title('Measures of Dispersion of African Countries against Year.')
    plt.ylabel('Value')
    plt.xlabel('Year')
    
create_plot(statistics.mean, 'Mean')
#create_plot(statistics.variance, 'Variance')
create_plot(statistics.stdev, 'Standard Deviation')
create_plot(interquartile_range, 'Interquartile Range')
create_plot(quartile_range, 'Quartile Range')
#create_plot(_range, 'Range')

plt.legend(loc='best')
```

Due to how large the variance and range are, i  would plot them separately

```python
create_plot(statistics.variance, 'Variance')
plt.legend(loc='best')
```

```python
create_plot(_range, 'Range')
plt.legend(loc='best')
```
